```python
def clear_punc(sent):
    sent = re.sub(r'[\W\dA-Za-z]','',sent)
    return(sent)
      
a = ""

print(clear_punc(a))

#stopword處理
stopwords = []
with open("stopwords.txt", "r", encoding="utf8") as file:
    for data in file.readlines():
        data = data.strip()
        stopwords.append(data)

bistopword = []
tristopword = []
quadstopword = []

for i in range(len(stopwords)):
    if len(stopwords[i]) == 2:
        bistopword.append(stopwords[i])
    elif len(stopwords[i]) == 3:
        tristopword.append(stopwords[i])
    elif len(stopwords[i]) == 4:
        quadstopword.append(stopwords[i])
        
```

```python
articles = pd.DataFrame(columns = ['title','content','time'])

listofwords=["大立光"]

for i in range(len(news)):
    count=0
    clear_title = clear_punc(str(news.title[i]))
    clear_article = clear_punc(str(news.content[i]))
    for k in range(len(listofwords)):
        if listofwords[k] in (clear_title):
            count+=1
        if listofwords[k] in (clear_article):
            count+=1
    if count>=1:
        articles = articles.append( [{'title': clear_title, 'time': news.post_time[i], 'content': clear_article}] )

articles
```

```python
listofdates = ['2016/01/01','2016/01/03']
datetimedates = [datetime.datetime.strptime(x,'%Y/%m/%d') for x in listofdates]


def selectdates(alldates, pdframe):
    result =  pd.DataFrame()
    for date1 in alldates:
        start_date = date1
        end_date = start_date + datetime.timedelta(days=1)
        mask = (pdframe['time'] > start_date) & (pdframe['time'] <= end_date)
        temp = pdframe.loc[mask]
        result =  pd.concat([result,temp])
    return(result)




subsetofarticles = selectdates(datetimedates, articles)
subsetofarticles.to_excel("C:\\Users\\carl_dis2003\\Desktop\\bigdata\\midterm\\outputs\\subset.xlsx")  
```
